```python
%matplotlib inline
import pickle
from matplotlib import pyplot as plt
import numpy as np
```

```python
def unpickle(file):
    with open(file,'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
def load_data(file):
    data = unpickle(file)
    
    raw_images = data[b'data']
    
    labels = np.array(data[b'labels'])
    
    images = convert_images(raw_images)
    
    return images, labels
```

```python
def convert_images(raw_images):
    
    raw = np.array(raw_images)
    
    images = raw.reshape([-1, 3, 32, 32])

    images = images.transpose([0, 2, 3, 1])

    return images
```

```python
images, label = load_data("data/data_batch_1")
```

```python
for i in range(10):
    print(label[i])
    plt.imshow(images[i], interpolation = 'nearest')
    plt.show()
```
